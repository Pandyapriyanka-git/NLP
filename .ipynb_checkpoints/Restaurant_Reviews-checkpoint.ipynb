```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Restaurant_Reviews.tsv', sep='\t')
```

```python
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
df['Liked'].value_counts()/len(df)
```

```python
like = df[df['Liked'] == 1]
dislike = df[df['Liked'] == 0]
```

```python
like.shape
dislike.shape
```

```python
data=pd.concat([like, dislike], axis=0, ignore_index=True)
```

```python
data.sample()
```

```python
data.shape
```

```python
data['Liked'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data['Review'], data['Liked'], test_size = 0.3, random_state =0, shuffle = True)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier 
from sklearn.pipeline import Pipeline
```

```python
classifier = Pipeline([("tfidf", TfidfVectorizer()) , ("classifier", RandomForestClassifier(n_estimators=100))])
```

```python
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.svm import SVC
```

```python
svm = Pipeline([("tfidf", TfidfVectorizer()) , ("classifier", SVC(C = 100, gamma='auto'))])
```

```python
svm.fit(X_train, y_train)
```

```python
y_pred = svm.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lgo = Pipeline([("tfidf", TfidfVectorizer()) , ("lgo", LogisticRegression())])
```

```python
lgo.fit(X_train, y_train)
```

```python
y_pred = lgo.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
model = Pipeline([("tfidf", TfidfVectorizer()) , ("model", MultinomialNB())])
model.fit(X_train, y_train)
print("Naive Bayes Accuracy:", model.score(X_test, y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
decision = Pipeline([
    ("tfidf", TfidfVectorizer()), 
    ("clf", DecisionTreeClassifier(random_state=62))
])
```

```python
decision.fit(X_train,y_train)
```

```python
y_pred = decision.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```
